```python
from langchain_community.document_loaders.csv_loader import CSVLoader
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.llms import HuggingFacePipeline

from langchain.chains import RetrievalQA
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
# Load CSV as documents
loader = CSVLoader(file_path="/kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")
documents = loader.load()
```

```python
# Embed documents using open-source model
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = Chroma.from_documents(documents, embedding_model)
```

```python
# Load local LLM
model_name = "tiiuae/falcon-7b-instruct"  # Or use mistral or llama models
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, trust_remote_code=True)
```

```python
# Wrap it in a pipeline for LangChain
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=256)
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
# Setup RAG chain
qa = RetrievalQA.from_chain_type(llm=llm, retriever=vectorstore.as_retriever())
```

# **Prompt 1: Add Formatting and Emphasis**

```python
query = """You are a business analyst AI.

Customer profile (below) has stopped using our services.

TASKS:
1. Identify the **top 2** reasons that likely caused this customer to churn.
2. Suggest **1 specific action** to prevent this type of churn in future.

📌 Please format your answer as:
- Reason 1:
- Reason 2:
- Recommendation:"""
```

```python
# Encourages structure and clear outputs.
result = qa.run(query)
print(result)
```

# **Prompt 2: Add Few-Shot Example**

```python
query = """You are a business analyst AI.

Example:
---
Customer:
Gender: Female, Contract: Month-to-month, InternetService: DSL, TechSupport: No

Reasons:
- Reason 1: Short-term contract increases churn risk
- Reason 2: Lack of technical support deters loyalty
- Recommendation: Offer yearly discounts and proactive support outreach
---

Now analyze the following customer profile and give reasons + a recommendation in the same format.
"""
```

```python
# Injects a sample so the LLM learns the pattern.
result = qa.run(query)
print(result)
```

# **Prompt 3: Chain of Thought Prompting**

```python
query = """You are a business analyst AI. A customer has churned.

First, reason step-by-step through their possible experience based on the data provided.

Then, identify the **top 2 causes** of churn and 1 **actionable recommendation** to prevent it.

Respond in bullet points."""
```

```python
# Encourages reasoning before conclusion.
result = qa.run(query)
print(result)
```

# **Prompt 4: Output Constraint Prompt (JSON Format)**

```python
query = """You are a business analyst AI. A customer has churned.

Identify the 2 most likely reasons and 1 recommendation.

Return the result in the following JSON format:
{
  "reason_1": "...",
  "reason_2": "...",
  "recommendation": "..."
}
"""
```

```python
# Encourages consistent, parseable structure.
result = qa.run(query)
print(result)
```

# **Prompt 5: Role + Goal Based Prompt** 

```python
query = """As a business intelligence system, your task is to reduce customer churn.

You are provided with a customer profile who recently stopped using our service.

→ Identify 2 main churn causes
→ Suggest 1 concrete, low-cost, high-impact action to retain such customers.

Respond briefly and clearly."""
```

```python
# Strong role + mission helps focus the output.
result = qa.run(query)
print(result)
```
